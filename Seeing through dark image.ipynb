```python
#Using thresholdin technique to read text from an image that was taken in very low ligth and not legible to human eyes


import cv2
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 20

#loading the image

dark_image = cv2.imread('dark_image.jpg',1)

plt.imshow(dark_image)
```

```python
## Convertin the image to grey scale

gray_image = cv2.cvtColor(dark_image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap = 'gray')
```

```python
## Checking the distribution of the values in the gray scale matrix
plt.hist(gray_image.ravel(),256)
plt.show()
```

```python
#Using thresholding technique to turn all pixels below a threshold to 0 and above it to 255
#After some trial and error, we get the value 8 

retval, threshold = cv2.threshold(gray_image, 8, 255, cv2.THRESH_BINARY)

plt.imshow(threshold,cmap = 'gray')
```

```python
#Using blurring techniques to smoothen the edges
#median blurring replaces the pixel values with the median of the values in the given kernel space

median = cv2.medianBlur(threshold,15)

plt.imshow(median,cmap = 'gray')
```

```python
## The image appears to be quite readable now.
# Lets compare it with what we had in the begining



plt.subplot(1,2,1).imshow(dark_image)
plt.title('Original Image',fontsize = 60, y =1.02)

plt.subplot(1,2,2).imshow(median,cmap = 'gray')
plt.title('After Transformations',fontsize = 60, y = 1.02)






```
